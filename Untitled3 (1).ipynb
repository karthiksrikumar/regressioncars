```python
df = pd.concat([selected_features, target], axis=1)

def plot_binned_scatter(x, y, xlabel, ylabel, title):
    # Create bins
    bins = np.linspace(x.min(), x.max(), 20)
    bin_centers = 0.5 * (bins[:-1] + bins[1:])
    
    # Bin the data
    digitized = np.digitize(x, bins)
    binned_means = [y[digitized == i].mean() for i in range(1, len(bins))]
    binned_counts = [len(y[digitized == i]) for i in range(1, len(bins))]
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.scatter(bin_centers, binned_means, s=np.array(binned_counts)*10, alpha=0.7, edgecolors='w', linewidth=0.5)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.grid(True)
    plt.show()

plot_binned_scatter(df['age'], df['car_amount'], 'Age', 'Car Purchase Amount', 'Binned Scatter Plot: Age vs Car Purchase Amount')
plot_binned_scatter(df['sal'], df['car_amount'], 'Annual Salary', 'Car Purchase Amount', 'Binned Scatter Plot: Annual Salary vs Car Purchase Amount')
plot_binned_scatter(df['net_worth'], df['car_amount'], 'Net Worth', 'Car Purchase Amount', 'Binned Scatter Plot: Net Worth vs Car Purchase Amount')
```
